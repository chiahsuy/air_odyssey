```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_flight_weather = pd.read_csv("flight_weather_all_final-va.csv")
```

```python
# simple visual to find histogram of delay
mask = (df_flight_weather['OriginState'] == 'VA') & (df_flight_weather['DepDelay'] > 0)
sns.histplot(df_flight_weather.loc[mask,'DepDelay'], stat='probability', kde=True)
plt.xlabel('Flight Departure Delays [minutes]');
```

```python
mask = (df_flight_weather['OriginState'] == 'VA') & (df_flight_weather['DepDelay'] > 0)
#df_flight_weather.loc[mask,'DepDelay'].hist(by=df_flight_weather['Operating_Airline'])
df_flight_weather.loc[mask,'DepDelay'].hist(by=df_flight_weather['IATA_Code_Marketing_Airline'], bins=100)
```
